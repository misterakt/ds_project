```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

# plot the histograms
plt.figure(figsize=(9,3))
plt.hist(x1, density=True, bins=20, alpha=0.5)
plt.hist(x2, density=True, bins=20, alpha=0.5)
plt.hist(x3, density=True, bins=20, alpha=0.5)
plt.hist(x4, density=True, bins=20, alpha=0.5);
plt.axis([-7,21,0,0.6])

plt.text(x1.mean()-1.5, 0.5, 'x1\nNormal')
plt.text(x2.mean()-1.5, 0.5, 'x2\nGamma')
plt.text(x3.mean()-1.5, 0.5, 'x3\nExponential')
plt.text(x4.mean()-1.5, 0.5, 'x4\nUniform')
```

```python
import numpy as np
from matplotlib.pylab import *
import matplotlib.animation as animation
#%matplotlib notebook

x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)


#n=100
def animate(curr):
    
    if curr == n: 
       a.event_source.stop()
    for ax in (ax1,ax2,ax3,ax4):
        ax.clear()
    ax1.hist(x1[:curr], density=True,bins=np.linspace(-10,1, num=50),alpha=0.5)
    ax2.hist(x2[:curr], density=True,bins=np.linspace(0,20,num=50),alpha=0.5)
    ax3.hist(x3[:curr], density=True,bins=np.linspace(5,25,num=50),alpha=0.5)
    ax4.hist(x4[:curr], density=True,bins=np.linspace(15,25,num=50),alpha=0.5)
    ax1.set_title('n={}'.format(curr))
    ax1.set_title('n={}'.format(curr))
    ax1.set_title('n={}'.format(curr))
    ax1.set_title('n={}'.format(curr))
    
a= animation.FuncAnimation(fig, animate, interval=1, repeat=False)

plt.show()
```
