```python
import pandas as pd
url_pop='https://en.wikipedia.org/wiki/Demographics_of_Germany'
url_ber='https://en.wikipedia.org/wiki/Demographics_of_Berlin'

df_pop=pd.read_html(url_pop,header=1)[5]
df_ber=pd.read_html(url_ber,header=0)[3]

def sub_string(year):
    cleaning=str(year).replace(" ¹","")
    return int(cleaning[-4:])

def make_it_number(num_string):
    return int(str(num_string).replace(" ",""))
    
df_pop=df_pop.rename(columns={"Unnamed: 0":"Year"})
df_pop=df_pop.loc[:118,['Year','Average population','Live births','Deaths']]
df_pop['Average population']=df_pop.apply(lambda x:make_it_number(x['Average population']),axis=1)
df_pop['Live births']=df_pop.apply(lambda x:make_it_number(x['Live births']),axis=1)
df_pop['Deaths']=df_pop.apply(lambda x:make_it_number(x['Deaths']),axis=1)
df_ber['Year']=df_ber.apply(lambda x:sub_string(x['Year']),axis=1)
```

```python
import matplotlib.pyplot as plt

#plt.plot('Year','Average population',data=df_pop,marker='',color='red',linewidth=2)
plt.plot('Year','Live births',data=df_pop,marker='',color='blue',linewidth=2,linestyle='dashed', label="birth")
plt.plot('Year','Deaths',data=df_pop,marker='',color='red',linewidth=2, label="death")
plt.legend()
plt.show
```

```python
#plt.plot('Year','Average population',data=df_pop,marker='',color='blue',linewidth=2,linestyle='dashed', label="Total pop")
plt.plot('Year','Population',data=df_ber,marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=2, label='Population in Berlin')
plt.plot('Year','Live births',data=df_pop,marker='',color='olive',linewidth=2,linestyle='dashed', label="birth in germany")
plt.plot('Year','Deaths',data=df_pop,marker='',color='red',linewidth=2,linestyle='dashed', label="Death in germany")
plt.legend()
plt.show
```

```python
fig, axs = plt.subplots(2,figsize=(15,15))
fig.suptitle('Berlin/Germany population change')
axs[0].plot('Year','Live births',data=df_pop,marker='',color='blue',linewidth=2,linestyle='dashed', label="birth")
axs[0].plot('Year','Deaths',data=df_pop,marker='',color='red',linewidth=2, label="death")
axs[0].set_title('Birth/death in population in Germany')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('number of population')
axs[0].legend()
axs[1].plot('Year','Population',data=df_ber,marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=2, label='Population in Berlin')
axs[1].plot('Year','Live births',data=df_pop,marker='',color='olive',linewidth=2,linestyle='dashed', label="birth in germany")
axs[1].plot('Year','Deaths',data=df_pop,marker='',color='red',linewidth=2,linestyle='dashed', label="Death in germany")
axs[1].set_title('Comparison population in berlin and Birth/death in population in Germany')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('number of population')
axs[1].legend()
```
